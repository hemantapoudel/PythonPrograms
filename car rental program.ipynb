Create a Python program that will compute the car rental prices.

The program takes the input of days, selection of VIP membership, and optional additions. The program calculates and displays the rental price, discount and final price of the car rental.

The basic price of car rental is $50 per day. The silver member gets 5% discount and gold member gets 10% discount of the total price. If the car is rented 7 days or more, 10% discount will apply.

The user should have the option of starting a new price calculation.

The program should include the validation check for missing or invalid data.

Please make use of functions wherever you can.

The prices of the additional items are listed below:

Optional Additions           Price/day

Excess Deduction                $20
Unlimited Kms                   $10
Additional Driver               $10 per rental

```python
days=int(input("Enter Days:"))
def Days_price(days):
    return 50*days
def Ex_Ded(days):
    return 20*days
def Un_Kms(days):
    return 10*days
def Add_Driver():
    return 10
price=Days_price(days)
print("Optional Additions:")
ex_ded=input("\tExcess Deduction(Y/N):").upper()
un_kms=input("\tUnlimited KMS(Y/N):").upper()
add_dvr=input("\tAdditional Driver(Y/N):").upper()
if(ex_ded=="Y"):
    price+=Ex_Ded(days)
elif(ex_ded=="N"):
    pass
else:
    print("Error in Excess Deduction")
if(un_kms=="Y"):
    price+=Un_Kms(days)
elif(un_kms=="N"):
    pass
else:
    print("Error in Unlimited Kms")
if(add_dvr=="Y"):
    price+=Add_Driver()
elif(add_dvr=="N"):
    pass
else:
    print("Error in Additional Driver")
print("Select MEMBERSHIP:")
membership=int(input("(1) Non-member, (2) Silver, (3) Gold:"))
if(days>=7):
    Discount=(10/100)*price
    Final_price=price-Discount  
elif(membership==1):
    Discount=0
    Final_price=price-Discount
elif(membership==2):
    Discount=(5/100)*price
    Final_price=price-Discount
elif(membership==3):
    Discount=(10/100)*price
    Final_price=price-Discount
else:
    print("Error in membership selection")
print("Car Rental Details:")
print("\tRental Price:",price)
print("\tDiscount:",Discount)
print("\tFinal Price:",Final_price)
```
